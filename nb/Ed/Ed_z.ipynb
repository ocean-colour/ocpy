# Examining $E_d$ vs. depth

```python
# imports
import numpy as np
from matplotlib import pyplot as plt

from ocpy.hydrolight import loisel23
from ocpy.utils import plotting
```

# Load up Loisel

```python
ds = loisel23.load_ds(1,0, profile=True)
```

```python
ds_IOP = loisel23.load_ds(1,0)
```

```python
ds
```

# Pick an example

```python
wave = ds.Lambda.data
wave.shape
```

```python
idx = 170
Edz = ds.Ed_z.data[:,idx,:]
Edz.shape
```

# Plot

## $a_{\rm nw}$

```python
fig = plt.figure(figsize=(8,5))
ax = plt.gca()
#
ax.plot(wave, ds_IOP.anw.values[idx])
#
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel(r'$a_{\rm nw}$ (m$^{-1}$)')
#
#ax.legend()
#ax.set_yscale('log')
#
plotting.set_fontsize(ax, 15)
#plt.savefig('Ed_vs_depth.png', dpi=300)
plt.show()
```

## $E_d$

```python
fig = plt.figure(figsize=(8,5))
ax = plt.gca()
#
for idepth in [0,5,10,20,30,60]:
    ax.plot(wave, Edz[idepth], label=f'z={ds.z.data[idepth]:0.1f}')
#
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel(r'Ed (W m$^{-2}$)')
#
ax.legend()
#ax.set_yscale('log')
plotting.set_fontsize(ax, 14.)
#
#plt.savefig('Ed_vs_depth.png', dpi=300)
plt.show()
```

## Relative

```python
fig = plt.figure(figsize=(8,5))
ax = plt.gca()
#
for idepth in [5,10,20,30,60]:
    ax.plot(wave, Edz[idepth]/Edz[0], label=f'z={ds.z.data[idepth]:0.1f}')
#
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel(r'Ed/Ed+ (W m$^{-2}$)')
#
ax.legend()
#ax.set_yscale('log')
plotting.set_fontsize(ax, 14.)
#
#plt.savefig('Ed_vs_depth.png', dpi=300)
plt.show()
```

## PAR

## $PAR \equiv \int_{400nm}^{700nm} \frac{E_0 \lambda}{hc} \, d\lambda$

```python
fig = plt.figure(figsize=(5,5))
ax = plt.gca()
#
ax.plot(wave, Ed*wave)
#
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel(r'$\lambda E_d$0+ (W m$^{-2}$)')
#
#ax.legend()
#ax.set_yscale('log')
#
#plt.savefig('Kd_components.png', dpi=300)
plt.show()
```

## Cumulative

```python
good_wv = (wave >= 400.) & (wave <= 700)
PAR_cum = np.cumsum((Ed*wave)[good_wv])
```

```python
fig = plt.figure(figsize=(5,5))
ax = plt.gca()
#
ax.plot(wave[good_wv], PAR_cum/PAR_cum[-1], 'k')
#
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Cumulative PAR')
#
#ax.legend()
#ax.set_yscale('log')
#
plt.savefig('PAR_cumulative.png', dpi=300)
plt.show()
```
