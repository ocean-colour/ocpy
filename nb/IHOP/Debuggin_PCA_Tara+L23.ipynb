# PCA Debuggin

```python
# imports
from importlib import reload

import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns

from sklearn import decomposition

from oceancolor.ihop import pca
```

```python
%matplotlib ipympl
```

# Run $N=5$

```python
reload(pca)
data, wave_grid, pca_fit = pca.generate_l23_tara_pca(return_N=5)
```

# Check the Eigenvectors

```python
fig = plt.figure(figsize=(6,6))
plt.clf()
ax = plt.gca()
#
for ss in range(5):
    ax.plot(wave_grid, pca_fit.components_[ss], label=f'{ss}')
#
ax.legend()
plt.show()
```

# Variance

```python
_, _, pca_fit_20 = pca.generate_l23_tara_pca(return_N=20)
```

```python
fig = plt.figure(figsize=(6,6))
plt.clf()
ax = plt.gca()
#
ax.plot(np.cumsum(pca_fit_20.explained_variance_ratio_), 'o')
#
plt.show()
```
